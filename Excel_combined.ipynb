This script would consolidate all tabs in an excel file and export the consolidated dataframe


```python
import pandas as pd

input_excel = pd.read_excel("/content/Lump Sum Payments_2025 and 2026 (1).xlsx",
                            sheet_name=None)
```

```python
tabs = input_excel.keys()
type(input_excel['UCD Med'])

consolidated_df = pd.DataFrame()

for tab in tabs:
  df = input_excel[tab]
  df['tab'] = tab
  consolidated_df = pd.concat([consolidated_df,df])

consolidated_df.shape
```
