The project is to find whether any employee reported in Unit 99 appeared in the UPTE juris list

```python
import pandas as pd

# Import the latest Unit 99 data
unit_99_data = pd.ExcelFile("/content/drive/MyDrive/Data/Unit_99_Title_Claim/Historical 99 Data/UPTE_99_as_of_6_3_2026.xlsx")
unit_99_data.sheet_names
```

```python
unit_99_df = unit_99_data.parse("Sheet1")
unit_99_df.shape
```

```python
unit_99_df.columns
```

```python
# function to standardize the email addresses for matching

def standardize_email(df, column_name):
    df[column_name] = df[column_name].str.lower().str.strip()
    return df
```

```python
# Import the latest juris_data

juris_data = pd.ExcelFile("/content/Combined_I254_DATA_HX_RX_TX_07062026.xlsx")
juris_data.sheet_names
```

```python
juris_df = juris_data.parse('All Units')
juris_df.shape
```

```python
juris_df.columns
```

```python
cols_to_keep = ['Employee (EE) ID', 'Employee Name', 'Bargaining Unit',
                'Location/Business Unit','Job Title Abbr Name', 'General Deductions Dues Indicator',
       'Annual Pay Rate 1.0 FTE', 'Step', 'Appt. %', 'Employee Class',
       'Employee Class Description', 'Appt Rep Code','Most Recent Date of Hire in Empl Record',
       'Earliest in Unit Begin Date for Empl Record', 'Job Expected End Date',
       'Department ID', 'Department ID Description', 'Location Description',
       'EE Work Location Building', 'EE Work Location Address1', 'EE Work Email Address',
       'EE Work Location Address2', 'EE Work Location Room/Suite',
       'EE Work Location Floor', 'EE Work Location City','EE Work Location State', 'EE Work Location Zip Code']
```

```python
# Save the updated juris data wit hselected column and standardized email addresses
juris_final_df = standardize_email(df=juris_df[cols_to_keep], column_name='EE Work Email Address')
juris_final_df.columns = [f"{col}_juris" if col != 'EE Work Email Address' else col for col in juris_final_df.columns ]
juris_final_df.head()
```

```python
# standardize 99 email address

final_99_df = standardize_email(df=unit_99_df, column_name='EE Work Email Address')
final_99_df.columns = [f"{col}_99" if col != 'EE Work Email Address' else col for col in final_99_df.columns ]
final_99_df.head()
```

```python
juris_in_99_email_match = juris_final_df.merge(final_99_df, how='inner', on='EE Work Email Address')
```

```python
juris_in_99_email_match.to_excel("June_99_in_juris_email_match.xlsx")
```
