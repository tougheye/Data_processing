```python
import pandas as pd
job_code_review_df = pd.read_csv("/content/Job Code Summary_1.13.26_review.csv")
job_code_review_df.tail()
```

```python
# drop the last 2 rows since they are not part of the data
drop_row_index = [ 63750, 63751]
job_code_review_clean_df = job_code_review_df.drop(drop_row_index)
# job_code_review_clean_df.tail()
job_code_review_clean_df.shape
```

```python
job_code_review_clean_df.columns
```

```python
job_code_review_clean_df.groupby('CTO/OS Subgroup')['Short Descr'].nunique().to_frame()
```

```python
job_code_review_clean_df['Short Descr'].value_counts()
```

```python
from numpy import astype
CTO_vs_BU_short_desc_crosstab = job_code_review_clean_df.pivot_table(values='Short Descr', index='CTO/OS Subgroup', columns= 'Union Code',
                                     aggfunc='nunique', fill_value=0, margins=True).astype(int)
CTO_vs_BU_short_desc_crosstab
```

```python
CTO_vs_BU_short_desc_crosstab.to_csv('CTO_vs_BU_short_desc_crosstab.csv')
```

```python
# cross tab to calculate row percentages
CTO_vs_BU_short_desc_crosstab_row_pct = CTO_vs_BU_short_desc_crosstab.div(CTO_vs_BU_short_desc_crosstab.sum(axis=1), axis=0) * 100
CTO_vs_BU_short_desc_crosstab_row_pct_rounded = CTO_vs_BU_short_desc_crosstab_row_pct.round(2)
CTO_vs_BU_short_desc_crosstab_row_pct_rounded
```

```python
CTO_vs_BU_short_desc_crosstab_row_pct_rounded.to_csv('CTO_vs_BU_short_desc_crosstab_row_pct_rounded.csv')
#
```

```python
CTO_vs_Short_desc_crosstb = pd.crosstab(index=job_code_review_clean_df['Short Descr'], columns=job_code_review_clean_df['CTO/OS Subgroup'], values= job_code_review_clean_df['Jobcode SETID'],margins=True, aggfunc='nunique')
```

```python
CTO_vs_Short_desc_crosstb
```

```python
CTO_vs_Short_desc_crosstb.to_csv('CTO_vs_Short_desc_crosstb.csv')
```

```python
import pandas as pd
job_codes_99_df = pd.read_csv("/content/Job Code Summary_1.13.26_non_represented_codes.csv")
job_codes_99_df.shape
```

```python
job_codes_99_df.columns
```

```python
job_codes_99_df.tail()
```

```python
drop_row_index = [ 23613, 23612]
job_codes_99_clean_df = job_codes_99_df.drop(drop_row_index)
```

```python
job_codes_99_clean_df.tail()
```

```python
# Get unique job codes & titles with necessary details
non_repr_code_details = job_codes_99_clean_df.groupby(['CTO/OS Subgroup',
       'CTO Description','Job Code','Job Code Description',  'Census/OCC Code',
       'Census/OCC Code Description','EEO1 Reporting Code',
       'EEO1 Reporting Cod Description']).size().reset_index(name='count')
non_repr_code_details.shape
```

```python
CTOs_outside_UPTE =['426', '436','456','577','621', '627', 'A20', 'A25', 'A30', 'A35', 'B30', 'Z25',
'C15', 'C25', 'F15', 'F25', 'F30', 'F35', 'F40', 'F45', 'G25', 'G35', 'G55', 'G65',
'G75', 'H10', 'I15', 'J10', 'J15', 'M05', 'M10', 'M20', 'S31', 'S56', 'TST', 'Z20']
```

```python
non_repr_code_w_short_des = non_repr_code_details.merge(job_code_review_clean_df[['Job Code', 'Short Descr']].drop_duplicates(), on='Job Code', how='left')
```

```python
target_99_titles_TCS = non_repr_code_w_short_des[non_repr_code_w_short_des['CTO/OS Subgroup'].isin(target_CTO)].reset_index(drop=True)
target_99_titles_TCS.shape
```

```python
target_99_titles_TCS.to_excel('target_99_titles_TCS.xlsx', index=False)
```

```python
non_repr_code_w_short_des.shape
```

```python
non_repr_code_w_short_des['Job Code'] = non_repr_code_w_short_des['Job Code'].astype(str).str.zfill(5)
non_repr_code_w_short_des['Census/OCC Code'] = non_repr_code_w_short_des['Census/OCC Code'].astype(str).str.zfill(6)
non_repr_code_w_short_des['EEO1 Reporting Code'] = non_repr_code_w_short_des['EEO1 Reporting Code'].astype(str).str.zfill(3)

non_repr_code_w_short_des.to_excel('non_repr_code_w_short_des.xlsx', index=False)
```

```python
Out_of_UPTE_titles_TCS = non_repr_code_w_short_des[non_repr_code_w_short_des['CTO/OS Subgroup'].isin(CTOs_outside_UPTE)].reset_index(drop=True)
Out_of_UPTE_titles_TCS.shape
```
