```python
import pandas as pd

full_99_list = pd.read_excel('/content/drive/MyDrive/UPTE-26-019_99 List_April 2026 (2) (2).xlsx', sheet_name="Report Page")
full_99_list.head()
```

```python
titles_of_interest = ['APPLICATIONS PROGR 1', 'APPLICATIONS PROGR 2', 'APPLICATIONS PROGR 3', 'APPLICATIONS PROGR 3 GF',
'BUS SYS ANL 1', 'BUS SYS ANL 2', 'BUS SYS ANL 3', 'BUS SYS ANL 3 GF', 'DATA SYS ANL 1', 'DATA SYS ANL 2', 'DATA SYS ANL 3',
'DATA SYS ANL 3 GF', 'INFO SYS ANL 1', 'INFO SYS ANL 2', 'INFO SYS ANL 3', 'INFO SYS ANL 3 GF','INSTRUCTIONAL DESIGNER 1',
'INSTRUCTIONAL DESIGNER 2', 'INSTRUCTIONAL DESIGNER 3', 'INSTRUCTIONAL DESIGNER 4', 'INSTRUCTIONAL DESIGNER 4 GF']
```

filter out the titles of interest from the main 99 list

```python
unit_99_titles_of_interest = full_99_list[full_99_list['Job Code Description'].isin(titles_of_interest)]
unit_99_titles_of_interest.head()
```

```python
unit_99_titles_of_interest.shape
```

Add the title groups to 99 list

```python
title_group_dict = {'APPLICATIONS PROGR 1 - 3': ['APPLICATIONS PROGR 1', 'APPLICATIONS PROGR 2', 'APPLICATIONS PROGR 3', 'APPLICATIONS PROGR 3 GF'],
                    'BUS SYS ANL 1 - 3': ['BUS SYS ANL 1', 'BUS SYS ANL 2', 'BUS SYS ANL 3', 'BUS SYS ANL 3 GF'],
                    'DATA SYS ANL 1 - 3': ['DATA SYS ANL 1', 'DATA SYS ANL 2', 'DATA SYS ANL 3', 'DATA SYS ANL 3 GF'],
                    'INFO SYS ANL 1 - 3': ['INFO SYS ANL 1', 'INFO SYS ANL 2', 'INFO SYS ANL 3', 'INFO SYS ANL 3 GF'],
                    'INSTRUCTIONAL DESIGNER 1 - 4': ['INSTRUCTIONAL DESIGNER 1', 'INSTRUCTIONAL DESIGNER 2', 'INSTRUCTIONAL DESIGNER 3', 'INSTRUCTIONAL DESIGNER 4', 'INSTRUCTIONAL DESIGNER 4 GF']}
```

```python
title_group_df = pd.Series(title_group_dict).explode().reset_index()
title_group_df.columns = ['Job Title Group', 'Job Code Description']
title_group_df.head()
```

```python
unit_99_titles_of_interest = unit_99_titles_of_interest.merge(title_group_df, on='Job Code Description', how='left')
unit_99_titles_of_interest.head()
```

```python
unit_99_titles_of_interest.to_excel('unit_99_titles_of_interest.xlsx')
```

create pivot tables with departments and job title groups

```python
unit_99_pivot_title_groups = unit_99_titles_of_interest.pivot_table(index=['Location/Business Unit', 'Department Description'],
                                       columns = 'Job Code Description',
                                       values = 'Employee Lived Name',
                                       aggfunc = 'count', fill_value=0)
unit_99_pivot_title_groups = unit_99_pivot_title_groups.reset_index()
```

```python
unit_99_pivot_title_groups.head()
```

importing the juris list that already includes the title group column

```python
TX_juris_df = pd.read_excel('/content/TX_I254_DATA_05042026_045959.xlsx', sheet_name='juris_list')
TX_juris_df.shape
```

```python
TX_pivot_table_dept_title_groups = TX_juris_df.pivot_table(index=['Location/Business Unit', 'Department ID Description'],
                                       columns = 'Job Title Group',
                                       values = 'Employee Name',
                                       aggfunc = 'count', fill_value=0,
                                       margins = True, margins_name='Total_TX_Cnt')

TX_pivot_table_dept_title_groups = TX_pivot_table_dept_title_groups.reset_index()
```

```python
tx_pivot_key_cols = ['Location/Business Unit', 'Department ID Description', 'Total_TX_Cnt']
tx_pivot_col_order =  tx_pivot_key_cols + [col for col in TX_pivot_table_dept_title_groups.columns if col not in tx_pivot_key_cols]
TX_pivot_table_dept_title_groups = TX_pivot_table_dept_title_groups[tx_pivot_col_order]
```

```python
TX_pivot_table_dept_title_groups.head()
```

Merge the 2 pivot tables keeping the unit 99 table as the main table

```python
# reset index to get the location and departments as regular columns
unit_99_pivot_title_groups.reset_index()
TX_pivot_table_dept_title_groups.reset_index()
```

```python
merged_df_all =  unit_99_pivot_title_groups.merge(TX_pivot_table_dept_title_groups,left_on=['Location/Business Unit', 'Department Description'],
                                 right_on = ['Location/Business Unit', 'Department ID Description'], how='left')
```

```python
merged_df_all.drop(columns=['Department ID Description'], inplace=True)
```

print the merged df into an excel with the tabs being the locations

```python
with pd.ExcelWriter('upte_99_title_groups__tx_cnts_universe.xlsx') as writer:

  for location, group in merged_df_all.groupby('Location/Business Unit'):
    sheet_name = location
    group.to_excel(writer, sheet_name=sheet_name, index=False)
#
```

Prepare the dataset keeping only the departments where UPTE members work

```python
merged_upte_depts_df = merged_df_all[~merged_df_all['Total_TX_Cnt'].isna()]
```

```python
with pd.ExcelWriter('upte_99_titles_tx_group_cnts_tx_depts.xlsx') as writer:

  for location, group in merged_upte_depts_df.groupby('Location/Business Unit'):
    sheet_name = location
    group.to_excel(writer, sheet_name=sheet_name, index=False)
```

```python
unit_99_target_dept_cnt = merged_df_all['Location/Business Unit'].value_counts().reset_index()
unit_99_target_dept_cnt.columns = ['Location/Business Unit', 'Total_99_Cnt']
```

```python
tx_dept_overlap_cnt = merged_upte_depts_df['Location/Business Unit'].value_counts().reset_index()
tx_dept_overlap_cnt.columns = ['Location/Business Unit', 'Total_TX_Cnt']
```

```python
dept_comparison_df = unit_99_target_dept_cnt.merge(tx_dept_overlap_cnt, on='Location/Business Unit', how='left')
```

```python
dept_comparison_df.to_csv('dept_comparison.csv')
```
