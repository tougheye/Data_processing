```python
import pandas as pd

upte_jan_df = pd.read_excel("/content/UPTE_Job_Summary_Represented_JobCode_01132026.xlsx")
upte_jan_df.shape
# (44017, 30)
```

```python
upte_jan_df.columns
```

```python
upte_mar_df = pd.read_excel("/content/UPTE_Job_Summary_Represented_JobCode_032026.xlsx", sheet_name='Represented Job Codes')
upte_mar_df.shape
# (46098, 30)
```

```python
upte_merged_df = upte_mar_df.merge(upte_jan_df[['Salary Plan SETID','Job Code','UCPATH Step', 'Sal Plan', 'Sal Plan Description','Grade', 'UC  Half Step','Hrly Rate', 'Monthly Rt', 'Annual Rt']],
                  on=['Salary Plan SETID','Job Code','UCPATH Step', 'Grade','UC  Half Step', 'Sal Plan', 'Sal Plan Description'],suffixes=('_mar', '_jan'), how='left')
upte_merged_df.shape
```

```python
upte_merged_df.columns
```

```python
cols_to_keep = ['Salary Plan SETID', 'Job Code','Job Code Description', 'UCPATH Step', 'UC  Half Step',
                'Hrly Rate_mar', 'Monthly Rt_mar','Annual Rt_mar', 'Hrly Rate_jan','Monthly Rt_jan',
                'Annual Rt_jan', 'Grade', 'Sal Plan', 'Sal Plan Description']
```

```python
upte_merged_df[cols_to_keep][upte_merged_df[cols_to_keep].duplicated()==True].sort_values(by=['Salary Plan SETID','Job Code','UCPATH Step', 'UC  Half Step'])
```

```python
upte_merged_df[(upte_merged_df['Salary Plan SETID'] == 'UCANR') & (upte_merged_df['Job Code'] == 4031.0)]
```

```python
upte_merged_df.sort_values(by=['Salary Plan SETID','Job Code','UCPATH Step', 'UC  Half Step'],inplace=True)
upte_merged_df[['Salary Plan SETID', 'Job Code',
       'Job Code Description', 'UCPATH Step', 'UC  Half Step', 'Hrly Rate_mar', 'Monthly Rt_mar',
       'Annual Rt_mar', 'Hrly Rate_jan',
       'Monthly Rt_jan', 'Annual Rt_jan']].head(100)
```

```python
upte_merged_df.columns
```

```python
final_comparison_df = upte_merged_df[cols_to_keep]
```

```python
final_comparison_df['Hrly_Inc%'] = (final_comparison_df['Hrly Rate_mar']-final_comparison_df['Hrly Rate_jan'])\
                                    /final_comparison_df['Hrly Rate_jan']
final_comparison_df['Monthly_Inc%'] = (final_comparison_df['Monthly Rt_mar']-final_comparison_df['Monthly Rt_jan'])\
                                    /final_comparison_df['Monthly Rt_jan']
final_comparison_df['Annual_Inc%'] = (final_comparison_df['Annual Rt_mar']-final_comparison_df['Annual Rt_jan'])\
                                    /final_comparison_df['Annual Rt_jan']
final_comparison_df.head(25)
```
