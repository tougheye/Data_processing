Upload Job Codes Summary file

```python
import pandas as pd

mar_2026_job_code_review_all = pd.read_excel("/content/drive/MyDrive/Data/UCOP_Job_Codes_Summary/UPTE-26-015_Job Summary Report_3.20.26.xlsx", sheet_name="Job Code Review")

# mar_2026_job_code_review_all.head()
```

UPload the Pivot summary of the 99 list to bring in the fields from Job Codes Summary. Then filter the titles that are in the CTOs where UPTE has representation

```python
head_cnt_all_99_df = pd.read_excel("/content/Apr_2026_99_list_summary.xlsx", sheet_name="All_Non-LBNL_99_title_cnts")
head_cnt_all_99_df.head()
```

```python
head_cnt_all_99_df.shape
```

Upload the UPTE CTO classes

```python
upte_classes = pd.read_excel("/content/Apr_2026_99_list_summary.xlsx", sheet_name="UPTE_classes")
upte_classes
```

```python
upte_cto_subgroups = upte_classes['CTO/OS Subgroup'].values
```

```python
upte_cto_subgroups
```

Separate Out Columns to keep from Job Code Review tab to join them on the 99 summary table

```python
job_code_review_cols_keep = ['Job Code', 'Descr', 'Short Descr', 'Union Code', 'FLSA Status']
```

Bring in the Census/OCC Code Description, EEO1 Reporting Cod Description, 'CTO/OS Subgroup', and CTO Description columns from Non-Represented Job Codes tab

```python
non_repr_job_codes = pd.read_excel("/content/drive/MyDrive/Data/UCOP_Job_Codes_Summary/UPTE-26-015_Job Summary Report_3.20.26.xlsx", sheet_name="Non-Represented Job Codes")
non_repr_job_codes.head()
```

```python
non_repr_job_codes.shape
```

```python
non_repr_job_code_cols_keep = ['Job Code Description', 'CTO/OS Subgroup',
       'CTO Description', 'EEO1 Reporting Cod Description', 'Census/OCC Code']
```

merge FLSA, short descriptions and job codes to the 99 summary

```python
head_cnt_all_99_w_jc_review = head_cnt_all_99_df.merge(mar_2026_job_code_review_all[job_code_review_cols_keep], left_on='Job Title', right_on='Descr', how='left').drop_duplicates(subset='Job Title')
```

```python
head_cnt_all_99_w_jc_review.columns
```

```python
head_cnt_all_99_w_non_rep_jc = head_cnt_all_99_w_jc_review.merge(non_repr_job_codes[non_repr_job_code_cols_keep], left_on='Job Title', right_on='Job Code Description', how='left').drop_duplicates(subset='Job Title')
```

```python
head_cnt_all_99_w_non_rep_jc.columns
```

```python
head_cnt_all_99_w_non_rep_jc.columns
```

Drop the duplicated Job code Description and Descr columns that are the same column as Job Title

```python
head_cnt_all_99_w_non_rep_jc.drop(columns=['Job Code Description', 'Descr'], inplace=True)
```

```python
final_non_lbnl_titles = head_cnt_all_99_w_non_rep_jc[head_cnt_all_99_w_non_rep_jc['CTO/OS Subgroup'].isin(upte_cto_subgroups)]
```

```python
column_df_list = final_non_lbnl_titles.columns
```

```python
column_order = ['Job Code','Job Title', 'Short Descr', 'UPTE Unit for CTO',
       'Current Status UPTE / UAW Titles in 99', 'Union Code', 'FLSA Status', 'CTO/OS Subgroup',
       'CTO Description', 'EEO1 Reporting Cod Description', 'Census/OCC Code', 'Grand Total',
       'Total All Others (Confirmed target)', 'All Others, Confidential',
       'All Others, Not Confidential', 'Excluded from Coverage',
       'Manager, Confidential', 'Manager, Not Confidential',
       'Not Applicable - Contingent Wk', 'Not Covered HEERA (Out of Stat',
       'Stud Academic Title, No HEERA', 'Supervisor, Confidential',
       'Supervisor, Not Confidential', 'Supervisor, Represented']
```

```python
final_non_lbnl_titles[column_order].to_excel("non_lbnl_apr_99_summary.xlsx", index=False)
```
